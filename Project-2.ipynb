# Problem 1 APR DRG Description and Total Charges

```python
import matplotlib.pyplot as plt
%matplotlib inline

## Foundation of Data Science: Project Problems
## Problem 1: How is the diagnosis description affecting the total charges of recieving treaments in Mount Sinai Hospital?

DandC = MountSinai[['APR DRG Description', 'Total Charges']]
DandC['Total Charges'] = DandC['Total Charges'].str.replace('$', '')
DandC['Total Charges'] = DandC['Total Charges'].astype(float)
Unique = DandC.groupby(['APR DRG Description']).agg({'Total Charges': ['sum', 'count']})
Unique['Average Charges'] = Unique['Total Charges']['sum']/Unique['Total Charges']['count']
Unique = Unique.drop(['Total Charges'], axis = 1)
Unique = Unique.sort_values(by = 'Average Charges', ascending = False)
BarChart = Unique.iloc[0:50].plot(kind = 'barh', title = 'Diagnosis vs. Total Charges', figsize = (50, 30)) 
# Compare the top 50 diagnosis that cost the most in Mount Sinai Hospital
plt.show()
```

The barchart for APR DRG Description and the Total Charges is plotted as seen above. The top 50 diagnosis were picked for plotting the chart. As we can seen from the results, Neonate w ECMO is the diagnosis generated the most averages.

```python
# load data, and change the data type into useabe one: 
Length = MountSinai['Length of Stay']
Charges = MountSinai['Total Charges']
Length = np.array(Length)
length = []
# modify the data to the way it can be process--> length of stay: remove the "+" sign 
for i in range(len(Length)):
    if Length[i] == '120 +':
        length.append('120')
    else:
        length.append(Length[i])
l = []
for j in range(len(length)):
    l.append(float(length[j]))

l = np.array(l)
print(l)
# modify the data to the way it can be process: change string into float --> length of stay: 
Charges = Charges.str.replace('$', '')
Charges = Charges.astype(float)
```

```python
from datascience import *
%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
import math
import numpy as np
from scipy import stats

def standard_units(any_numbers):
    "Convert any array of numbers to standard units."
    return (any_numbers - np.mean(any_numbers))/np.std(any_numbers)

def correlation(t, label_x, label_y):
    return np.mean(standard_units(t.column(label_x))*standard_units(t.column(label_y)))

def slope(t, label_x, label_y):
    r = correlation(t, label_x, label_y)
    return r*np.std(t.column(label_y))/np.std(t.column(label_x))

def intercept(t, label_x, label_y):
    return np.mean(t.column(label_y)) - slope(t, label_x, label_y)*np.mean(t.column(label_x))

def fit(table, x, y):
    """Return the height of the regression line at each x value."""
    a = slope(table, x, y)
    b = intercept(table, x, y)
    return a * table.column(x) + b
```
